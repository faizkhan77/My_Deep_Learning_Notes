### True Positive and False Positive
So, we have this dataset of Dogs images where 6 images are of dogs and the other 4 are not dog, so there is a Binary Classification here, Dog vs Not a dog (1 or 0)


Lets say u build a ML Model, or lets say u askl someone to make a Guess of what these images are, and lets say this is your predictions

This is a Prediction, hence it wont be true and would make some mistakes, so lets think Prediction as your base and lets only talk about Positive Predictions which is Dog (1), so we will gray out or ignore the "No Dog" predictions for now

And now u have a total of 7 dog predictions, now OUT OF THOSE 7, HOW MANY OF THEM ARE CORRECT? Well lets mark the predictions as correct or not correct for your Positive class (1).

So out of these 7, whichever are the CORRECT predictions are called **True Positive**, so here the "Positive" means your class in binary, meaning dog as dog is 1 and not dog is 0.

So out of all our Positive predictions, how many are true/correct? so to identify this u simply compare it with reality/truth, so in our case out of all the Positive predictions we only got 4 **True Positive**

Similarly, out of the 7 Positive predictions, 3 were wrong, so these are called **False Positive**


So keep in mind the "Positive" word indicates the class in Binary Classification, meaning 1

### True Negative and False Negative

Similarly, here the **Negative** Means the 0 class which is Not a dog.

Okay, so now think about No Dog predictions ang ignore the Dog prediction

Now, here two of the Predictions are wrong and one is right, so here, So of the 3 Negative predictions only 1 is right, so **True Negative** is 1 and out of the 3 Negative predictions 2 are wrong so **False Negative** is 2


## Accuracy
So below we have marked all our Predictions

 Basically these shows, out of all our predictions, which are correct and which are wrong, 
 Now, how many we got right? welp we got 5 right prediction
 
 And that is called **Accuracy**
 
 
 So it means out of all your predictions, how many you got right, doesnt matter Positive or Negative class, so we got 5 out of 10 right, hence **Acuracy** is 5/10 = **0.5**

## Precision

Now going back to the only **Positive** class

Here we saw **True Positive** = 4 and **False Positive** = 3.

So **Precision** is out of all **dog predictions**, how many you got right? 

formula = `True Positive / Total num of Predictions for Positive class`

In our case = `4 / 7 = 0.57`

So hence, the **Precision for Dog** class(Positive) is **0.57**.



Similary, the **Precision for No Dog** class(Negative) is `True Negative / Total num of predictions for Negative class` meaning `1 / 3 = 0.33`

## Recall

When ur thinking about **Recall**, u always think about Truth as ur baseline, 

so whats our Truth for lets say the Dog class(Positive)? Well we got total 6 True images of dogs so yh truth is 6 and out of the 6th dog class truth, how many did we predicted correctly? meaning whats our **True Positive**? well its 4

So, **Recall** is out of all your dog class truth, how many did u predicted right?

Formula = `True Positive / Num of Truth of the Positive class`

In our case = `4 / 6 = 0.67`

Similary, the Recall for No Dog class(Negative) is `True Negative /  Num of Truth of the Negative class` meaning `1 / 4 = 0.25`

## Precision vs Recall

So Precision and Recall has a certain difference 

Which is when youre thinking about **Precision**, always think about **Predictions** as your base

When youre thinking about **Recall**, think about **Truth** as your base



By the way, Precision and Recall is for individual class

## F1-SCORE

F1-Score formula is `2 * (precision - recall) / (precision + recall)`

So this just gives u the overall health, or overall performance of your model
