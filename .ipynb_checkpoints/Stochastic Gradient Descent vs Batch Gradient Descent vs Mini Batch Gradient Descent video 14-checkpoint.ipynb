Now, we will try to understand the differences between **Stochastic Gradient Descent**, **Batch Gradient Descent** and **Mini Batch Gradient Descent**

## Batch Gradient Descent

Lets say, we are building a ML model for House price prediction

Here u have the Training dataset which is area, bedrooms, price and we are trying to build the model which is nothing but just a Linear Equation which can tell you the Price based on area and bedrooms

In this Linear Equation all we are trying to do is find the optimal value for w1, w2 and bias

Our dataset have total 6 examples

1. During ML Model training, what we do is, we first initialize a random value for the weights and bias

2. Then we go and feed the first sample into the Equation and find the predicted price and compare it with the actualprice and find out the error, the error here is the Squared error


3. Then we feed the second sample, find predicted price, compare with actual then find the error, we do this for all the Sample

4. Once we did the process for all the samples, we completed **One Epoch** and at the End of an Epoch, we simply Sum all the errors then take the Average of it, which is the **Mean Squared Error (MSE)**


5. After that, we adjust and update the weights and bias using the Derivatives


6. So, above lets say our weights are 51 and 9 and bias is 20001

7. Now, we use that same weights and bias and go through all the Training samples again for another epoch

So we keep on running this process for n number of epochs untill our error gets to a minimum level, and at that time we will have the optimal value of weights and bias

And then we can say our ML model has been trained!

So this entire process of going through entire Trainng sample to complete one epoch and update the weights and bias is called **Batch Gradient Descent**

In **Batch Gradient Descent** we go through all the Training samples before we adjusting the weights


### Concerns

But, what if we have 10 millions samples instead of 6? To find the Total error we will have to run the epoch and every epoch will go through 10 Million samples to find 10 Million errors then sum them up then it will also have to calculate Derivative of 10 Million weights and bias to adjusts the parameters

And if you have more features, lets say in this case we have 2 features (age, bedrooms) then it will be 10 Million each, meaning it will need to find 20 Million Derivatives, now what if you have like 200 features?? That would be an insane amount of Computation!!

## Stochastic Gradient Descent

1. So what we can do is, instead of going through all the samples, we go to 1 Randomly picked up sample, find out the error

2. And after 1 sample itself, you start adjusting the weights and bias, so u find the Derivative then adjust the parameters

3. And then Again u randomly pick 1 sample, find error

4. Again u adjust weights and bias


So, here, instead of going through all the samples, we adjust the weights and bias after just every random sample

So this approach or method is called **Stochastic Gradient Descent (SGD)**

## Batch vs SGD Key Differences

They key difference between the two is **Batch Gradient Descent** uses all the Training sample before it start making adjusmment to the parameters and  **Stochastic Gradient Descent (SGD)** use only 1 randomly picked up sample then adjusts the parameters

**Batch Gradient Descent** is good for small dataset since it can have high Computation

 **Stochastic Gradient Descent (SGD)** is good for very big dataset as its light
 
The Graph between cost and epoch looks something like below for both of them where **Batch Gradient Descent** is much smoother and for  **Stochastic Gradient Descent (SGD)** its more of zig zag



## Mini Batch Gradient Descent

**Mini Batch Gradient Descent** is quite similar to **SGD**, in **SGD** we use only 1 random sample,

But in **Mini Batch Gradient Descent** we use a BATCH of randomly picked training sample, because while doing **SGD**, if you have lets say 10 Million samples and if u find errors for each of the sample one by one then it might be slow, and not only slow, it cant use the vectors mathemetics of numpy, see vector math is one of the big advantages of why people use numpy, u cant make use of that

So in 10 million samples, lets say we can take a Batch of 20 random samples each and train the model

Or lets say we have 20 Training samples, we can pick our batch size to be 5 means we will randomly pick 5 samples for each batch, do forward pass and then find the error for each batch and then adjusts the parameters

## Batch Gradient Descent Implementation

1. The batch gradient will take X and y and also epochs and learning_rate as input parameters
2. We get the num of features
3. Then we initialize weights and bias to be 1 and 0 using numpy
4. Then we get total samples by using len() function
5. Now we will use for loop to go through the epochs one by one
6. Now, we need to calculate y_predicted using the Linear equation so we will do it simply using numpy dot() method and T for transpose (its to convert row into columns). dot() method will multiply each value of weights with X one by one
7. Now from partial derivative equation we need to find out derivative of erros with respect to weights and bias

8. Now we need to update the weights and bias
9. Now, we gotta find the Cost/Loss using Mean Squared Error (MSE), so first of all we need to sum all our Squared error then get the Average, u can do it easily using numpy by doing np.mean(np.squared(y - y_predicted))
10. Now, we saw the graph above, lets plot it, so we will need list of all the costs and epochs, so we will create an empty list and append the costs and epochs but for every 10 iterations only, not all
11. Then lastly, we will just return the weights and bias and also cost and epoch list since we wanna plot


```python
import numpy as np
def batch_gradientdescent(X, y, epochs, learning_rate=0.01):
    num_of_features = X.shape[1]
    w = np.ones(num_of_features)
    b = 0
    n = len(X)
    
    epoch_list = []
    cost_list = []
    
    for i in range(epochs):
        # calculate linear equation
        y_predicted = np.dot(w, X.T) + b
        
        # calculate gradients (derivative)
        w_gradient = -(2 / n) * (X.T.dot(y - y_predicted))
        b_gradient = -(2 / n) * np.sum(y - y_predicted)
        
        # update weights and bias
        w = w - learning_rate * w_gradient
        b = b - learning_rate * b_gradient
        
        # get the cost
        cost = np.mean(np.square(y - y_predicted))
        
        # update weights and bias
        if i % 10 == 0:
            epoch_list.append(i)
            cost_list.append(cost)
            
    return w, b, cost, epoch_list, cost_list
```

```python
w, b, cost, epoch_list, cost_list = batch_gradientdescent(X_scaled, 
                                                          y_scaled.reshape(y_scaled.shape[0], ), 500)
w, b, cost
```

Now using this weight and bias we can now make our Prediction function

The thing is we needed to scale our feature everytime we pass it during prediction or trainng, we dont wanna do that, we will create such that it will scale it automatically inside the prediction function

1. We will take area, bedroom, weight and bias as input param
2. We need to transform area and bedroom by scaling it, so we will simply use MinMaxScaler
3. Then we will use Linear Equation to calculate the price, this will be scaled automatically since we are using the scaled X to calculate price
4. Then we will need the actual value of price, not the scaled one, so we will simply inverse the scaling using **inverse_transform()** method of MinMaxScaler

```python
def predict(area, bedrooms, w, b):
    scaled_X = x_scaler.transform([[area, bedrooms]])[0]
    
    scaled_price = w[0] * scaled_X[0] + w[1] * scaled_X[1] + b
    return y_scaler.inverse_transform([[scaled_price]])[0][0]

predict(2600, 4, w, b)
```

## SGD Implementation

SGD Implementation is the same as Batch GD, its just that instead of going through every sample, we will chose only 1 random sample for every epoch and thats all!, the rest is same, Using Linear equation to predict then calculating the gradient (derivative) then updating the parameters

But how do you randomly chose a single value? well we can use the **randint()** from **random** module and the range will be from 1 to n-1, so this will work as an index of the samples then we will simply pick random value for X and y using the random index

```python
import random
def sgd(X, y, epochs, learning_rate=0.01):
    num_of_features = X.shape[1]
    w = np.ones(num_of_features)
    b = 0
    n = len(X)
    
    epoch_list = []
    cost_list = []
    
    for i in range(epochs):
        random_index = random.randint(0, n-1)
        random_X = X[random_index]
        random_y = y[random_index]
        
        y_predict = np.dot(w, random_X.T) + b
        
        w_gradient = -(2/n) * (random_X.T.dot(random_y - y_predict))
        b_gradient = -(2/n) * (random_y - y_predict)
        
        w = w - learning_rate * w_gradient
        b = b - learning_rate * b_gradient
        
        cost = np.square(random_y - y_predict)
        
        # update weights and bias
        if i % 100 == 0:
            epoch_list.append(i)
            cost_list.append(cost)
            
    return w, b, cost, cost_list, epoch_list
```
