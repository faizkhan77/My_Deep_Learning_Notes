### Generic Function to plot samples

So we can kind of see that first one is a Frog and second is a Truck, anyways, since we will be Plotting the image again and again, lets just create a Generic function to do it, it will take X, y and index as parameters, and also we want to print the label or class as well so we can know which image is it, so we will just create a List and store all the Classes, and they are in Order, for example 0 means Airplane, 1 is automobile and 2 is bird, etc

```python
classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
```

```python
y_train[:2]
```

We can see above the first 2 samples is 6, and 6 is Frog, and second sample is 9 means Truck.

One thing we wanna do is Reshape the array as u can see above it gives as 2D array, but we want only the value directly, no array, just a 0D array, so we will reshape it.

The way you reshape is below, so the first dimension which is 50k, we dont wanna reshape or change that, so if u wanna keep it as it is just pass **-1**, and then the second dimension is u wana Flatten the array, so for example instead of getting [6], we want only 6, so for this, we will just leave it blank

```python
y_train = y_train.reshape(-1,)
y_train[:3]
```

So u can see, instead of a 2D array, its a 1D array, now when you check a particular value u will get the value directly as 0D

```python
y_train[0]
```

Now lets make the Generic function

```python
def plot_sample(X, y, index):
    plt.figure(figsize=(10,2))
    plt.imshow(X[index])
    plt.title(classes[y[index]])
```

```python
plot_sample(X_train, y_train, 0)
```

```python
plot_sample(X_train, y_train, 1)
```

```python
plot_sample(X_train, y_train, 100)
```

See, it worked perfectly!

### Normalization (Divide pixels by 255)

So this was a Quick Data Exploration part we just did, Now we want to Normalize our data, we saw in previopus tutorial that whenever u have an IMage, u want to scale it by dividng each Pixels value with 255 which is the max RGB channel, So if u divide it by 255, u will be Normalizing it into 0 to 1 range, because u can see the values here are not scaled

```python
X_train[0]
```

We wanna bring it to 0 to 1 range, so we can simply divide by 255

```python
X_train[0] / 255
```

U can see they are in 0 to 1 range after u divide each value with 255, so will do the same for all Training and Testing samples

```python
X_train = X_train / 255
X_test = X_test / 255
```

### Artificial Neural Network (ANN)

Now, we will build a simple **ANN** first so we can see the performance of how **ANN** works and then will do **CNN** and see how the performance improve.

In previous tutorial in the GPU Benchmark video we have used this same dataset and over there we used **ANN**, so i will just copy that same code

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(32,32,3)),
    keras.layers.Dense(3000, activation="relu"),
    keras.layers.Dense(1000, activation="relu"),
    keras.layers.Dense(10, activation="sigmoid")
])
model.compile(optimizer="SGD", 
             loss="sparse_categorical_crossentropy",
             metrics=['accuracy'])
```

In the GPU Benchmark tutorial we used "categorical_crossentropy" as loss, but here we will use "sparse_categorical_crossentropy". 

Here a good pic to understand the difference between them

So wheever you have your Y as OneHotEncoded vector where you have multiple output and only the predicted feature is 1 rest is 0, for example above we can see that we got bunch of 0 and the predicted value is **Ship** so only ship is 1, rest 0, we did this same thing on the GPU Benchmark tutorial by applying OneHotEncoding on the y_train and y_test, so anyways, if you have a OneHotEncoded Y like this then use **"categorical_crossentropy"** as loss.

Use **"sparse_categorical_crossentropy"** when your Y is directly a value (Only 1), like in the pic above its 8 as ship is number 8.

```python
model.fit(X_train, y_train, epochs=5)
```

Well anyways, so after Training, we can see that we got very very low Accuracy,only **49%** even during the GPU Benchmark video if u remember we got low accuracy as well using **ANN**,

Yes i know we did only 5 epochs, but even if you do more epochs the Accuracy would still be low, i wont do more than 5 since i dont have a GPU and it will take forever to Train 

Lets Evaluate it on Test samples

```python
model.evaluate(X_test, y_test)
```

So u can see its pretty low, so basically **ANN** is performing really bad on this dataset with 5 Epochs, lets print the Classification report as well and see, **Numpy's argmax** will simply give us the Index of the max value in each output of y_pred

```python
from sklearn.metrics import classification_report
import numpy as np
y_pred = model.predict(X_test)
y_pred_class = [np.argmax(element) for element in y_pred]

print(classification_report(y_test, y_pred_class))
```

We can see the Precision and Recall on each of the classes is pretty bad and also the F1-Score is low as well, so overall **ANN** doesnt perform well when it comes to complex Image Classification

### Convolutional Neural Network (CNN)

In the previous video about **CNN** we saw that **CNN** will have couple of layers, usually these layers will be **Convolutional layer + ReLU** and then **Pooling** layer for Dimensionality reduction, then again **Convolutional layer + ReLU** and then **Pooling** and then you have a **Dense Network**


So, we will keep the same **Dense Network**, all we will do is, before the **Dense Network**, will add some **Convolutional layer + ReLU** and **Pooling** layers.

1) So when ur in the middle layer, we dont need to specify the shape because it the Network can figure it out itself

2) Also, to keep things simple we will only keep one Hidden layer in the **Dense Network** as **CNN** layers would have done Most of the work already, so we wont need so many Deep layers.

3) For the Output, we will be using a **Softmax** Activation function, its very Popular, will see the difference later

4) So anyways, now in the **CNN** Layers, we will first add a **Convolutional** Layer using  **"layers.Conv2D"**. Now,**Convolutional** is used to detect Features, now we saw that form Previous video tutorial that we saw that for image 9, we can have this kind of 3 Filters, so the first one is Detecting the loopy pattern, second one is Detecting the Vertical edge and third one is Detecting the Diagonal pattern

The best thing about **CNN** is, u dont have to tell it What the **Filters** are, it will figure out the Filters for u, U only need to tell the Filters size and how many Filters u want. So in the case of 9 we got 3 Filters and 3 Features Maps

    1) But in our case now, we will just use random, hmm u know maybe just 32 Filters, so this mean, it can Detect 32 Different Features or edges in your Image, 
    2) And then **"kernel_size"** is the Filter size, so we can say that we want a 3 by 3 Filters, even in the Digit 9 example above we have 3x3 Filters, so thats what **"kernel_size"** is.
    3) Then we will specify Activation which is **ReLU**
    4) Then input shape will be the shape of the each image sample, which is 32 by 32 with 3 RGB Channel
    
<br> 

5) Then we will add the **Max Pooling** Layer using **"layers.MaxPooling2D"**
    1) In the parameter we will specify the Pooling size as 2 by 2

<br>

Now you can have one set of Convolutional and Pooling layer or Multiple sets, u can figure it out by Trial and Error, so just for fun, we will add another set of Convolutional and Pooling layer 

```python
cnn = keras.Sequential([
    # CNN Layers
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation="relu", input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation="relu"),
    layers.MaxPooling2D((2, 2)),
    
    # Dense Network
    layers.Flatten(),
    layers.Dense(64, activation="relu"),
    layers.Dense(10, activation="softmax")
])
```

Now we will simply do our usual Model Compile thing and then Train

```python
cnn.compile(optimizer="adam", 
             loss="sparse_categorical_crossentropy",
             metrics=['accuracy'])
```

```python
cnn.fit(X_train, y_train, epochs=10)
```

We can see that after just 10 epochs it gave us around **80%** accuracy, actually if u compared it with the **ANN** we did above, here u can see just after 5 Epochs we got around **73%** accuracy whereas in **ANN** we got **49%**.

So u can see that using **CNN** helps u Tremendously!

Now lets Evaluate on Test set

```python
cnn.evaluate(X_test, y_test)
```

And we got 70% accuracy which is pretty good, we can add more Convolutional and Pooling set or increase the Epochs to get more accuracy as well.

But u see for images like this, which is kind of hard for even us to tell which image is which since this is actually a Tough dataset, so even in this kind of complex tough dataset also we got a pretty decent accuracy with just 10 epochs.

So good job, we just build our first ever **CNN**!!!

Now we will do Plotting of some samples in the Test set using the same Generic func we build above, but first just like we reshaped y_train above to 1D arrays, we have to do with y_test as well

```python
y_test = y_test.reshape(-1,)
plot_sample(X_test, y_test, 1)
```

Now lets just predict all Test dataset and we will get y_predicted value, then we will apply **Numpy's Argmax** to get the index of the max probability, since it will return u probability of each class due to **Softmax** func in the Output layer

```python
y_pred = cnn.predict(X_test)
y_pred_class = [np.argmax(i) for i in y_pred]
```

Lets check how it predicted the first 5 samplees

We can also Plot the prediction 

```python
for i in range(10):
    plot_sample(X_test, y_test, i)
    plt.xlabel(f"Predicted: {classes[y_pred_class[i]]}")
```

Now lets check the Classification report

```python
print(classification_report(y_test, y_pred_class))
```

See? We got much better numbers now in our Classification report by using **CNN** as compared to **ANN**!!
