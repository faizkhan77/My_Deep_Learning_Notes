### Example 1
Suppose u are a CEO of Apple, and ur given some sheet of Revenues and Expenses of Iphone,Ipad and Mac on 2018,19 and 2020 and you are told to caclulate the profit and loss, we already know how to do it its simple, all you do is substract Expenses from Profit

### Example 2
Lets take another case, Here we have Iphone,Ipad and Mac and num of units sold per year (2018,19,20) and another sheet containing Price per unit for each Iphone,Ipad and Mac,

### Matrix Multiplication
Now you want to find out how much Total sales we did every year, lets say as a CEO, u wanna find out how much sales u have done in 2018, so what u can do is, multiply first 1000 with 30, that will give you total Sales of Iphone, then multiply 400 with 5 for Ipad and 1200 with 2 for Mac 

and this way u can get a Total amout of sales every year.

So u can do the same with 2019 and 2020 and u will get a formula like this



So here, what we basicaly did is **Matrix Multiplication**


So the first grid matrix is price per unit for each products and second matrix is num of unit sold in each of the year, 

you simply multiply the first matrix first row with each column in second matrix and that gives you the result

### Dot Product
So you multiply 1000 with 30, 400 with 5 and 1200 with 2 then we add all the multplication up

So you do the same thing for second and third column, so this thing is called **Dot Product**


So **Dot product** is used in  **Matrix Multiplication**, it is nothing but, Multiplying two vectors, 

so when you have single row and single column they are called **Vector**

and when you do that kind of Multiplication, they are called **Dot Product**

#### Example 2
Now, to do the same thing for the above example, num of unit sold example of Iphone, Ipad and Mac, and we want to find the Total revenue from each year

```python
price_per_unit = np.array([1000,400,1200])
units = np.array([[30,40,50],[5,10,15],[2,5,7]])
```

```python
print(price_per_unit)
print(units)
```

Now, if u did simple muliplicatio by the way, it will do soething like this

```python
price_per_unit*units
```

#### Broadcasting in Numpy
So it basucally Multiplied 1000 by 30, 400 by 40 and 1200 by 50 (multiplying price_per_unit row with each row in units matrix) so if the shape of 2 matrix is not same, so it will try to expand it meaning like this:

    [1000,400,1200], [1000,400,1200], [1000,400,1200]
    [30,40,50],  [5,10,15],   [2,5,7]

So it repeated/expanded [1000,400,1200] so its the same shape as the second matrix and then it multiply each rows, so this concept of expanding a smaller shaped array in numpy as called as **Broadcasting** in Numpy

#### Dot product in Numpy
But this is not what we want, we want the sum of each output columns after multiplying like this


So this Operation is called **Dot product** and Numpy has a func called dot() which will return you the multiplication and also sum each column up

```python
ans = np.dot(price_per_unit, units)
ans
```

So 34400 shows the sales amount in the year 2018, 50000 for 2019 and 64400 for 2020!.

So we can see Matrix is pretty simple, its just an arrangement of numbers in rows and colums!
