# Simple Neural Network with no Hidden layers
Now we wil create a Simple Neural Network, where it just has 2 layers, input layers with 784 elements and output layers with 10 elements/neurons.



Ad the way we create this in Tensorflow and Keras is, you use keras.Sequential(), sequential basically mean im having a stack of layers in my Neural Network, and since its a stack it wil accept every layer as one element

So first element inside Sequential(), now, Keras has this API of keras.layers.Dense, Dense means every neurons is connected to every other neurons in the second layer, so we will create a Dense layer where it takes 3 params, output shape is 10 cuz we got 10 classes 0 to 9, and input shape is 784,basically we are specifying input and output layer here, then we have to specify the Activation func where in this case we will use Sigmoid, thats it, we defined our simple neural network

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation="sigmoid")
])
```

Now we need to Compile, in Tensorflow and Keras, this is something u have to do all the time where u kinda compile the neural network and u pass bunch of arguments

so first one is optimizer, we will use adam optimizer, u might have lots of question what is adam optimizer? well hold ur curiosity cuz we will cover it later, but optimizer basically allows u to Train efficiently when the backward propagation and the training is going on, optimizer will allow u to reach to global optima in a efficient way,

the second param is loss, which is ur loss function, we will just use "sparse_categorical_crossentropy", what this means is, our Output class are categorical cuz we have categories, 0 to 9, and sparse means out output variables is actually an integr number since 0-9 is an integer, u can use MSE as well

third param is metrics and we will put accuracy, basically mean my goal is to make my model accurate ao thats my metrics

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

Now we can use fit method where the Training actually happens, so will supply our training set and epoch, means num of itertions, now when we run it, it will perform the Training, during the training u can see that, in each epoch/iteraion, it is going through these samples 1875 samples,  and it is measuring the loss and the accuracy, and u will notice that, as the time goes, the accuracy imcreases

```python
model.fit(X_train_flattened, y_train, epochs=5)
```

Lets try to improve the accuracy by Scaling our values, often in ML, u need to scale the values, so how do we scale them? well we know that our input is within the range of 0 to 255, so if we divide the whole Array by 255 it will be scaled from 0 to 1

```python
X_train = X_train / 255
X_test = X_test / 255
```

Then we will flatten it again

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
X_train_flattened[0]
```

Now, we can see that its scaled, Scaling is a technique that can improve the accuracy of the model, so will train again

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

We can see that we got some improvements in our Accuracies, in our 5th iteration we got 92% accuracy

Now lets try to Evaluate our model on our Test dataset cuz before deploying ur model u always need to evaluate ur model on a Test dataset and below is how you do it

```python
model.evaluate(X_test_flattened, y_test)
```

Well we got 92% which is the same and its pretty decent enough!

Its a simple Neural Network, it takes less time for Training

Now, lets do Test sample predict just for fun

# Confusion matrix in Tensorflow

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

# Using hidden layer

Now we will create the same Neural network again but this time with a Hidden layer.

So as u add hidden layers, it generally tends to improve the performance,

so in original layer, we can say we want 100 neurons (trial and error to find best num of neurons), then the same input_shape and in activation function we will use "relu" this time

then in the hidden layer we want 10 neurons and "sigmoid" activation, the rest is same

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

we can see now with a hidden layer our accuracy is like 98% which is an insanse improveme,

lets evaluate test samples

```python
model.evaluate(X_test_flattened,y_test)
```

Then will plot the same confusion matrix again

```python
y_predicted = model.predict(X_test_flattened)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

We can see alot of improvements now from this matrix as compared to the previus one

Now, one thing i dont like is whenever we build a neural network, u have to always make a flattened array, but luckily if u dont want to make those flatten array, keras comes up with A special layer called flatten, so lets say u dont wanna fit X_train_flattened but directly X_train.

Simply add a new line above called keras.flattened() where u pass in the size which is 28 x 28 then in the second line we can remove input size cuz it can figure it out frm the first line, so this makes it simple, this way u dont have to create a flattened array

# Using Flatten layer so that we don't have to call .reshape on input dataset

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10)
```

We can see as we increased the epoch to 10, acuracy is 99% which is crazy!

```python
model.evaluate(X_test,y_test)
```
