## More Cost functions
So we have already looked at MAE, but there are other types of erros such as  MSE

MSE is same as MAE, only difference is, instead of taking the Sum of Absulute errors, we take the Sum of Square of the Absulute values in MSE which allows your **Gradient Descent** to converge in a better way.

The third type of error we have is **Log Loss** or **Binary cross entropy**, this one is also widely used by people

## Python codes for manual implementation
Now we will implement all these Cost functions in Python manually, the reason is, Cost func is used in NN Training, so in future tutorial we will implement NN from scratch and it will be important for us to understand **Log Loss** cuz will use **Logistic Regression** and will be using **Log Loss**

So we build to numpy arrays, one is predicted and other is actual value.

```python
import numpy as np
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

### MAE Implementation
First we will implement MAE, there wil be 2 parameters which is true and predicted, now as we show from above presentation, MAE is nothing but the Absolute difference between predicted and truth and then u simply take an Average of that, so lets write the code

```python
def mae(y_true, y_predicted):
    total_errors = 0
    for truth, predicted in zip(y_true, y_predicted): # zip() to iterate through 2 arrays at once
        total_errors += abs(truth - predicted) # abs() will give u the Absolute difference
    print(f"Total Erros is {total_errors}")
    
    mae = total_errors / len(y_true) # will give the Average which is our MAE
    print(f"MAE is {mae}")
```

```python
mae(y_true, y_predicted)
```

So, as we can see above, it worked pretty great! Now, we have used for loop above, but the reason Numpy is very powerful is because it allows u to do Vector operations very very easily, so we could do the same thing above but in 1 line using Numpy

```python
np.mean(np.abs(y_true - y_predicted))
```

See? We did it in one line, we can see how powerful really Numpy is!!!

If u wanna get Total Error, u can use sum() instead of mean()

```python
np.sum(np.abs(y_true - y_predicted))
```

### Log Loss / Binary Cross Entropy manual Implementation

Before we implement the Log loss, u need to know a certain things

When we do np.log([0]) it says not defined

```python
np.log([0])
```

So, whenever u have log([0]), what we need to do is, we need to take a value which is very close to 0 but not 0 because we cant use 0 since it says undefined, so for example if we do lets say 0.000000000000001 it would work as its slightly more than 0 but not 0

```python
np.log([0.000000000000001])
```

So in our Log loss formula, we have to replace 0 with a value that is very closee to 0 but not 0.

So first of all we are taking only log of Ypredicted not Absolute value

so we are taking log of Ypredicted + (1 - truth value) which means if you have the value 1 in Ypredicted that will create a problem because 1 - 1 would be log([0]) which is undefined.

So first thing we need to do is, come up with Epsilon, so lets say our Epsilon is 1e raise to -15, this simply mean after decimal u will have 15 zeros and then 1

```python
epsilon = 1e-15
```

```python
y_predicted
```

See, what we are trying to do is, if we have 1 in y_predicted, we wanna replace it with a value which is close to 1 but not 1, like 9.9999999

Similarly, we wanna replace 0 with a value close to 0 but not zero which is like 0.000000000001.

So we will use List Comprehension to go over all value in y_predicted, so here each i will be each value of y_predicted, and we simply want to take a maximum between i and epsilon, so it wil take the Max value means 1, and leave the min value like 0 with 1e-15

```python
y_predicted_new = [max(i, epsilon) for i in y_predicted]
y_predicted_new
```

Then will do the same thing to take the Min value meaning 0 as it is but leave the max meaning 1 as 0.999999999999999999999

```python
y_predicted_new = [min(i, 1-epsilon) for i in y_predicted_new]
y_predicted_new
```

We have to do all of these just becayse log of 0 is undefined.

Now u can simply do the log of y_predicted_new u get these

```python
np.log(y_predicted_new)
```

if we have done the log of the orignal y_predicted, we would get these values which is not good, thats why we needed to do all of that above

```python
np.log(y_predicted)
```

Now here, we can put together the formula of Log Loss


```python
# Clipping Predicted Values: The np.clip function ensures that y_predicted_new values are
# within a valid range to avoid issues with logarithms of 0.
# Note : line no.4 is not there in Codebasics tutorial, i added it by seeing in ChatGPT as it
# wasnt working for me
y_predicted_new = np.clip(y_predicted_new, epsilon, 1 - epsilon)

-np.mean(y_true * np.log(y_predicted_new) + (1-y_true) * np.log(1 - y_predicted_new) )
```

## Log Loss (in functions)

So below, we simply put together everything we did above for LogLoss inside a function

```python
def logloss(y_truth, y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i, epsilon) for i in y_predicted]
    y_predicted_new = [min(i, 1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.clip(y_predicted_new, epsilon, 1 - epsilon)

    return -np.mean(y_truth * np.log(y_predicted_new) + (1-y_truth) * np.log(1 - y_predicted_new) )
```

```python
logloss(y_true, y_predicted)
```
