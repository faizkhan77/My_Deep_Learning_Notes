## Valid Convolution

If you remember from our last tutorial on CNN, in that we discuss the Hand written digit recognization and we came up with 3 Different **Filters** or Feature detectors for digit 9, they are loopy detector, vertical line detector at middle and the tail which is diagonal line detector


And when we apply this Filter, what we find is, we reduce the input size by a little bit, so in this case, the input was 5 by 7 and when you apply the 3 by 3 Filter, you get 3 by 5 **Feature Map** which is an Output


And the Formula is: `(m - f + 1) x (n - f + 1)` where
* m and n is the Dimension of the Original image input which is 5x7
* f is the Filter dimension which is 3


So `(5 - 3 + 1) x (7 - 3 + 1) = 3 x 5`, so we got 3 by 5, So this is called a **Valid Convolution** or a **Valid Padding**

## Solution using Padding (Same Convolution)

To solve this problem what we can do is, we can Pad the Original Image with lets say here, in this case we are using 1 by 1 Padding


So on each side of your grid, we are adding 1 row or 1 column of empty pixels like above, which is called **Padding**, but you can put some values like -1, -1 basically represents a blank background, so u can put -1 in all the Padding, i only put in the top left corner, but it would be in all the padding


So, Now when we Apply 3x3 Filter, we can start from very corner

And we can go like this...




So this way, even the corner pixels like this -1 here, it will play alot of role, 

because see, it will also Contribute in this **Convolution Operation**


And when this Filter moves down, it will still play a role in Multiple times while detecting this features.

<br>

Now, if u think about the Dimensions, the Original Image was 5x7, but we added 2 columns (left and right) and 2 rows (up and down), so it becomes 7x9 with **Padding 1**


And if u do the Math using the same Formula `(m - f + 1) x (n - f + 1)` means

`(7 - 3 + 1) x (9 - 3 + 1)` You will realize that, you will get `5 x 7` which is the the Original Image, So Original Image was 5x7, we applied **Padding** of 1, then we Apply **Filter** then we get 5 by 7 back, So we are basically Preserving the Image back and at the same time the Corner pixels gets to Play a Fair Role in the Feature Detection.


So this is called **Same Convolution**


## Valid Convolution and Same Convolution

So, just to Summarize, we have **Valid Convolution** which means no **Padding**.
**Same Convolution** means Pad such that the Output is SAME as the input (Original Image)


## Valid and Same Convolution in Tensorflow

If you are using **Tensorflow**, often you will find this argument of **padding**


Where you will either supply value of **"same"** or **"valid"**, so if u come across this argument, now u know whats **Valid Convolution** vs **Same Convolution**
