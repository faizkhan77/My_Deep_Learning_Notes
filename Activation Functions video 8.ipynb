# Vanishing Gradient Problem
Now the issue with Sigmoid and tanh is this, if u know about Derivative, Derivative is nothing but, like similar to slope, when u have non linear func at every point the slope is changing so then you use Derivative to express that slope

So Derivative is nothing but Delta y divide by Delta x, which its telling u how much an output changes for a given change in input, so for example for our insurance data, it will say how much a person would likely buy an insurance,how much that will change based on how the Age changes.

Now once u have derivative, what happen for Sigmoid and tanh is, think about the higher values where the line is getting flat, lets say ur value is 4, so here whats delta  y? y is 1 for 3 and y is 1 for 4 in x-axis so basically there is no changes in y, so the change in y is 0 and change in x is 1, and 0/1 is 0. 

same on the negative range as well where there is no changes and this can create a problem in ur learning process cuz we saw in the ML Gradient descent tutorial that we need to calculate the Derivative and backpropagate ur errors and if ur derivative is closing towards 0 the Learning becomes extremely slow, this is called **Vanishing Gradient Problem**


So Sigmoid and tanh has this **Vanishing Gradient Problem** and that makes the learning rate super slow

# ReLU

So that is why we use another Activation function called **Rectified Linear Unit (ReLU)**



This is a very simple function where if your value is more than 0 (positvie) then it return the value as it is, else if the value is less than 0 (negative) it will always return 0.

So example if i feed 2 as an input it will return as it is, 2

But if i feed -2 it will return 0 since its negative, its a very simple function

The guideline is if ur not sure what to use then always use ReLU for hidden layer activation function


because if u think about the math behind ReLU, it is computationally very effective, in Sigmoid and tanh u have to do some compuitation but in ReLU its very simple and lightweight and thayts why its very very popular.

Now, ReLU has this problem called **Dead neurons problem** meaning, since ReLU returns 0 for negative values, and if its 0, it means the neurons are dead and doesnt work

# Leaky ReLU
Due to that there is another flavour of ReLU called the **Leaky ReLU**

Where there is still some sort of linear line, it is still trying to reduce the value close to 0 but the equation now instead of 0 is 0.1

# PYTHON CODE FOR EACH ACTIVATION FUNCTION

### Sigmoid function

formila : z = 1 / 1 + e^-z

where e is euler's number = 2.71828

Output will be between 0 and 1

```python
import math
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
sigmoid(2)
```

```python
sigmoid(100)
```

```python
sigmoid(-100)
```

```python
sigmoid(34)
```

```python
sigmoid(-100)
```

### tanh function

formila : z = e^z - e^-z / e^z + e^-z

where e is euler's number = 2.71828

Output will be between 1 and -1

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(12)
```

```python
tanh(-23)
```

```python
tanh(34)
```

```python
tanh(4)
```

### ReLU function

formila : z = max(0, x)

This one is very simple cuz we are just taking max value between 0 and x

Value more than 0 will be returned as it is and value below 0 wil always be 0

```python
def relu(x):
    return max(0, x)
```

```python
relu(21)
```

```python
relu(-21)
```

### Leaky ReLU function

formila : z = max(0.1x, x)

This one is also very simple, it will not make negative value 0 cuz its multiplying 0.1 with x, so sligtly away from 0

```python
def leaky_relu(x):
    return max(0.1 * x, x)
```

```python
leaky_relu(10)
```

```python
leaky_relu(-3)
```
