We will use Home price prediction example because its the most simplest one but **Chain Rule** applied to all kind of Deep Learning NN Problems

So we simply do this same process and go through all the samples in the Training dataset 1 by 1 till the very end and we then accumulate all these errors and we take the Mean out of it, its called **MSE**, also called **Loss**

So once u have your loss, you Backpropagate.

## How Chain Rule works

So, going back to our simple architecture here, whats happening is, we are taking a Loss, then we are taking a **Partial Derivatie of loss with respect to e**

Now if u remember the power rule of Derivatives, The Derivatives of loss with respect to e (which is an error) will be **2e**, because we put the <sup>2</sup> in front of e then we substarct 1 from <sup>2</sup> like this **2e<sup>1-2</sup> = 2e**

And then we take **Partial Derivative of error with respect to ŷ** (y_predicted) and the answer is **-1** because in Partial Derivative we discard or ignore the other variable, in this case **y**, so **ŷ** doesnt have any value ontop of it so it means its 1, so u bring that 1 infront of **ŷ** and u get **-1** 


Then we take the **Partial Derivative of ŷ (y_predicted) with respect to weights**

The answer is **area** because again, in Partial Derivative we ignore the other variable making them 0 except **w1**, so we are just left with **w1 * area** and the Partial Derivative of that with respect to w1 would be **area**


Now, what is the **Partial Derivative of Loss with respect to w1**??


We got those individual partial derivatives above now we wanna get Partial Derivative of Loss with respect to w1, but why do we do that? well because Partial Derivative is all about **for a given change in w1 how much my Loss would change**, so when we are doing NN Training, we would wanna know when we feed inputs, we wanna know for a given change in w1 what is the change in loss? because we are trying to come up with Optimal value of w1 and w2 and Partial Derivative helps in that process


So, this would be nothing but Multiply all those Partial Derivatives above and you get **Partial Derivative of Loss with respect to w1**

And if u studied math u do know that this two "de" will cancel out and this two "dŷ" will also cancel out

So eventually what we will get is "dl/dw1"


So mathematically it make sense, So all we are doing is taking the Partial Derivative in each steps then Backward propagating then simply multiplying all the partial derivatives to get  **Partial Derivative of Loss with respect to w1**, same with w2 and bias

So here, our final  **Partial Derivative of Loss with respect to w1** will be **-2e \* area** because we are multplying **2e \* -1 \* area** so **2e - 1** is **-2e \* area** 



So this equation of Backpropagating is called the **CHAIN RULE**, its very simple, we are just Chaining our Partial Derivatives multiplying them, basicaly we are taking products of our Partial Derivatives and this product is nothing but a **CHAIN RULE**
